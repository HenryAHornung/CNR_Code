```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
data = pd.read_csv("data.csv", sep = ";", index_col = "Unnamed: 0")
data = data.drop("chords", axis = 1)
data = data.drop("time_signature", axis = 1)
data = data.drop("normalized chords", axis = 1)
data = data.drop("parsed chords", axis = 1)
data = data.drop("normalized and parsed chords", axis = 1)
data["year"] = None
data.head()
```

```python
# Loading the landing page of the website
jazzstandards_all = requests.get("https://www.jazzstandards.com/compositions/").text
jazzstandards_all = BeautifulSoup(jazzstandards_all, "html.parser")
```

```python
# Getting links to each other list 
indeces = ["00", "01", "01-b", "02", "02-b", "02-c", "02-d", "03", "03-b", "03-c", "04", "04-b", "05", "06"]
js_links = []
for index in indeces:
    js_links.append(f"https://www.jazzstandards.com/compositions/index-{index}.htm")
```

```python
# Using the list of links to extract the name and year of publication of each song.
js_names = []
js_year = []

for link in js_links:
    js_page = requests.get(link).text
    js_page = BeautifulSoup(js_page, "html.parser")
    js_list = js_page.find_all("tr", {"class": "JSContentsLine"})
    for js in js_list:
        js_names.append(js.find("a").text)
        js_year.append(js.find_all("td")[1].text)
```

```python
for i, name in enumerate(js_names):
    name = name.replace(":", ".")
    js_names[i] = name.lower()
```

```python
# Matching the song names to the years in a dictionary.
js_dict = {}
for i, name in enumerate(js_names):
    if js_year[i] != "Trad.":
        js_dict[name] = js_year[i]
```

```python
for i, name1 in enumerate(data["song"]):
    for name2 in js_dict.keys():
        if name1.lower().strip(" ") == name2.lower().strip(" "):
            data.at[i, "year"] = js_dict[name2]
```
